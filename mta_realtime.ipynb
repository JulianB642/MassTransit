```python
import gtfs_realtime_pb2
import requests
requests.__version__
```

```python
#initialize an instance of FeedMessage
#FeedMessage is a class in the google library
feed=gtfs_realtime_pb2.FeedMessage()
url='https://api-endpoint.mta.info/Dataservice/mtagtfsfeeds/nyct%2Fgtfs'
key='1vdKwxazQ45e0xMSzyW2Sa0cQjPSk5ps23uqrden'
#dictionary of HTTP headers
headers={'x-api-key':key}
response=requests.get(url,headers=headers)
```

```python
#pass response to the parser in FeedMessage object
feed.ParseFromString(response.content)
```

```python
#the feed is a gtfs_realtime_pb2.FeedMessage object
print(type(feed))
#Taking a look at the feed, it's composed of nested
#dictionaries
#There is a header dictionary, followed by a dictionary
#of entities
#Entities are indexed by id
#Not every entity is 
feed
```

```python
trip_updates={}
i=0
for entity in feed.entity:
    if entity.HasField('trip_update'):
        trip_updates[i]=entity.trip_update
        i+=1
```
